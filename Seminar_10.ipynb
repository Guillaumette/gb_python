**Задание 6:** Создать новый столбец в таблице с пингвинами, который будет отвечать за показатель длины клюва пингвина (flipper_length_mm).


high - длинный (от 42)

middle - средний (от 35 до 42)

low - маленький (до 35)

```python
penguins.loc[penguins['bill_length_mm'] >= 42, 'beak_length'] = 'high'
penguins.loc[(penguins['bill_length_mm'] >= 35) & (penguins['bill_length_mm'] < 42), 'beak_length'] = 'middle'
penguins.loc[penguins['bill_length_mm'] < 35, 'beak_length'] = 'low'
penguins
```

**Задание 7:** Изобразить гистограмму по flipper_length_mm с оттенком beak_length. Сделать анализ



```python
sns.histplot(data=penguins, x='flipper_length_mm', hue='beak_length')
```
